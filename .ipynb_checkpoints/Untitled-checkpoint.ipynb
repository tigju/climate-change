```python
from statsmodels.graphics.tsaplots import plot_acf
for country in grouped.groups:
    c = grouped.get_group(country)
    plot_acf(c['Temperature'])
    plt.title(country)
    plt.show()
```

```python
target = pd.DataFrame()
cols = ['Temperature', 'Precipitation']
results = pd.DataFrame()
for country in grouped.groups:
    c = grouped.get_group(country)
    for col in cols:
        temp = c.rename(columns={"Year": "ds", 'Country': "Country", col:'y'})
        print(temp)
        m = Prophet()
        # fit the data
        m.fit(temp)
        # make forecast for next two years
        future = m.make_future_dataframe(periods=12*2, freq='Y')
        forecast = m.predict(future)
        #     m.plot(forecast)
        # rename column , so it won't be ovewriten by next city
        forecast = forecast.rename(columns={'yhat': f'{country}_{col}'})
    df = pd.concat(forecast, axis=1)
    target = pd.merge(target, df.set_index('ds'), how='outer', left_index=True, right_index=True)
    
```
